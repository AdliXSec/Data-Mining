# Data Preprocessing

Dari dataset yang dikirimkan, tahapan pembersihan & transformasi yang diminta adalah:
1. **Missing value handling** pada variabel `Leukosit` yang dibedakan berdasarkan atribut `Category` (0 & 1). Menggunakan nilai *Mean* (rata-rata).
2. Ekspresi **Standarisasi (Z-Score)** untuk parameter `Leukosit`.
3. **Transformasi** variabel `Sex` yang berisi teks menjadi numerik (Label Encoding).

```python
import pandas as pd
import numpy as np
from io import StringIO

df = pd.read_csv("Quiz Live Coding I(Sheet1).csv", sep=';', decimal=',')

print("Tampilan Awal Dataset:")
display(df.head(10))
```

### 1. Missing Value Handling (Variabel Leukosit)

```python
print("Jumlah missing value pada Leukosit sebelum imputation:", df['Leukosit'].isna().sum())

# Menghitung rata-rata Leukosit per Kategori
mean_leukosit = df.groupby('Category')['Leukosit'].mean()
print("Rata-rata Leukosit per Kategori:")
print(mean_leukosit)

# Mengisi missing value dengan rata-rata berdasarkan masing-masing kategori
df['Leukosit'] = df.groupby('Category')['Leukosit'].transform(lambda x: x.fillna(x.mean()))

print("Jumlah missing value pada Leukosit setelah imputation:", df['Leukosit'].isna().sum())
```

### 2. Standarisasi Variabel Leukosit Menggunakan Z-Score

```python
# Rumus Z-Score: (X - Mean) / Standard Deviation
mean_l = df['Leukosit'].mean()
std_l = df['Leukosit'].std()

df['Leukosit_ZScore'] = (df['Leukosit'] - mean_l) / std_l

print("Hasil Standarisasi Z-Score pada variabel Leukosit:")
display(df[['Category', 'Leukosit', 'Leukosit_ZScore']].head(15))
```

### 3. Transformasi Variabel Sex

Karena variabel `Sex` di dataset semuanya merepresentasikan "P", kita mengubah data alfabetik *string* ini menjadi skalar numerik. Oleh karena itu kita asumsikan P = 1. (Jika terdapat `L` pada populasi sesungguhnya, dapat diubah menjadi 0).

```python
df['Sex_numeric'] = df['Sex'].map({'P': 1})

print("Hasil Transformasi Variabel Sex:")
display(df[['Sex', 'Sex_numeric']].head())

print("Dataset Akhir:")
display(df.head(15))
```
