```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.decomposition import PCA

print("MEMUAT DATASET")
df = pd.read_csv('DATASET_TA1(Sheet1).csv')
print(f"Bentuk data awal: {df.shape} baris & kolom")

num_cols = ['Age', 'Hemoglobin', 'Hematokrit', 'Lekosit', 'Eritrosit', 'Trombosit', 'HbA1c', 'RBG']
```

```python
print("SOAL 1: CLEANING KESELURUHAN (TANPA KLASIFIKASI)")
# Membuat copy agar tidak mengganggu soal 2 & 3
df_step1 = df.copy() 
print("Jumlah Missing Value SEBELUM (Keseluruhan):")
print(df_step1.isnull().sum()[df_step1.isnull().sum() > 0]) # Hanya print yang ada null

# Imputasi keseluruhan dengan median (numerik) dan modus (kategori)
df_step1[num_cols] = df_step1[num_cols].fillna(df_step1[num_cols].median())
df_step1['Gender'] = df_step1['Gender'].fillna(df_step1['Gender'].mode()[0])

print("\nJumlah Missing Value SESUDAH (Keseluruhan):")
print(df_step1.isnull().sum())
```

```python
print("SOAL 2: CLEANING KHUSUS DM TYPE 2")
mask_dm2 = df['Diagnose'] == 'DM TYPE 2'
print("Missing value SEBELUM pada DM TYPE 2:")
print(df[mask_dm2].isnull().sum()[df[mask_dm2].isnull().sum() > 0])

# Imputasi numerik dengan median dari kelompok DM TYPE 2
for col in num_cols:
    median_val = df.loc[mask_dm2, col].median()
    df.loc[mask_dm2, col] = df.loc[mask_dm2, col].fillna(median_val)

# Imputasi kategori (Gender) dengan modus dari kelompok DM TYPE 2
mode_val_dm2 = df.loc[mask_dm2, 'Gender'].mode()[0]
df.loc[mask_dm2, 'Gender'] = df.loc[mask_dm2, 'Gender'].fillna(mode_val_dm2)

print("\nMissing value SESUDAH pada DM TYPE 2:")
print(df[mask_dm2].isnull().sum().sum(), "Total Missing Value")
```

```python
print("SOAL 3: CLEANING KHUSUS DM TYPE 2 + PENYERTA")
mask_penyerta = df['Diagnose'] == 'DM TYPE 2 + Penyerta'
print("Missing value SEBELUM pada DM TYPE 2 + Penyerta:")
print(df[mask_penyerta].isnull().sum()[df[mask_penyerta].isnull().sum() > 0])

# Imputasi numerik dengan median dari kelompok DM TYPE 2 + Penyerta
for col in num_cols:
    median_val = df.loc[mask_penyerta, col].median()
    df.loc[mask_penyerta, col] = df.loc[mask_penyerta, col].fillna(median_val)

# Imputasi kategori (Gender) dengan modus dari kelompok DM TYPE 2 + Penyerta
mode_val_penyerta = df.loc[mask_penyerta, 'Gender'].mode()[0]
df.loc[mask_penyerta, 'Gender'] = df.loc[mask_penyerta, 'Gender'].fillna(mode_val_penyerta)

print("\nMissing value SESUDAH pada DM TYPE 2 + Penyerta:")
print(df[mask_penyerta].isnull().sum().sum(), "Total Missing Value")
```

```python
print("SOAL 4: DETEKSI DAN HAPUS OUTLIER")
print(f"Jumlah baris SEBELUM hapus outlier: {df.shape[0]}")

def remove_outliers(data, cols):
    data_clean = data.copy()
    for col in cols:
        Q1 = data_clean[col].quantile(0.25)
        Q3 = data_clean[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        data_clean = data_clean[(data_clean[col] >= lower_bound) & (data_clean[col] <= upper_bound)]
    return data_clean

df = remove_outliers(df, num_cols)
print(f"Jumlah baris SESUDAH hapus outlier: {df.shape[0]}")
```

```python
print("SOAL 5: CEK DAN HAPUS DATA DUPLIKAT")
print(f"Jumlah duplikat SEBELUM dihapus: {df.duplicated().sum()}")
print(f"Jumlah baris SEBELUM hapus duplikat: {df.shape[0]}")

df = df.drop_duplicates()

print(f"Jumlah duplikat SESUDAH dihapus: {df.duplicated().sum()}")
print(f"Jumlah baris SESUDAH hapus duplikat: {df.shape[0]}")
```

```python
print("SOAL 6: PELABELAN DIAGNOSE & GENDER")
print("SEBELUM Pelabelan:")
print(df[['Gender', 'Diagnose']].head(5))

df['Diagnose'] = df['Diagnose'].map({'DM TYPE 2': 0, 'DM TYPE 2 + Penyerta': 1})
df['Gender'] = df['Gender'].map({'L': 0, 'P': 1})

print("\nSESUDAH Pelabelan (L=0, P=1 | DM2=0, DM2+Penyerta=1):")
print(df[['Gender', 'Diagnose']].head(5))
```

```python
print("SOAL 7: STANDARISASI MIN-MAX SCALER")
print("SEBELUM Min-Max Scaler:")
print(df[['HbA1c', 'RBG']].head(5))

minmax = MinMaxScaler()
df[['HbA1c', 'RBG']] = minmax.fit_transform(df[['HbA1c', 'RBG']])

print("\nSESUDAH Min-Max Scaler (Range 0-1):")
print(df[['HbA1c', 'RBG']].head(5))
```

```python
print("SOAL 8: STANDARISASI Z-SCORE")
z_cols = ['Hemoglobin', 'Hematokrit', 'Lekosit', 'Eritrosit', 'Trombosit']
print("SEBELUM Z-Score (Nilai asli):")
print(df[z_cols].head(5))

scaler = StandardScaler()
df[z_cols] = scaler.fit_transform(df[z_cols])

print("\nSESUDAH Z-Score (Mean ~0, StdDev ~1):")
print(df[z_cols].head(5))
```

```python
print("SOAL 9: DATA REDUCTION (PCA)")
print(f"SEBELUM PCA: Terdapat {len(z_cols)} variabel Z-Score ({z_cols})")

pca = PCA(n_components=2) # Reduksi menjadi 2 komponen utama
pca_result = pca.fit_transform(df[z_cols])

df['PCA1'] = pca_result[:, 0]
df['PCA2'] = pca_result[:, 1]

print("\nSESUDAH PCA: Menghasilkan 2 variabel komponen utama (PCA1 & PCA2):")
print(df[['PCA1', 'PCA2']].head(5))
```

```python
print("SOAL 10: FEATURE SELECTION")
print(f"Semua kolom yang tersedia: \n{df.columns.tolist()}")

# 1. Menambahkan Uji Korelasi untuk memvalidasi Feature Selection
correlation = df.corr()
print("\nKorelasi variabel terhadap Diagnose (diurutkan):")
print(correlation['Diagnose'].sort_values(ascending=False))

df_selected = df[['Gender', 'Age', 'Diagnose', 'HbA1c', 'RBG']]

print("Dataset Akhir setelah Feature Selection:")
print(df_selected.head(5))
```
