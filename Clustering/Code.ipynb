```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# 1. LOAD DATA 
df = pd.read_csv('DATASET_TA1(Sheet1).csv', sep=';', decimal=',')

# 2. PEMILIHAN VARIABEL
selected_features = ['Hemoglobin', 'Hematokrit', 'HbA1c', 'RBG']
data_cluster = df[selected_features].copy()
```

```python
# 3. DATA PREPARATION (Imputasi & Standarisasi)
data_cluster = data_cluster.fillna(data_cluster.median())

scaler = StandardScaler()
data_scaled = scaler.fit_transform(data_cluster)
```

```python
# 4. K-MEANS CLUSTERING (k=2)
kmeans = KMeans(n_clusters=2, random_state=42, n_init=10)
df['Cluster'] = kmeans.fit_predict(data_scaled)
```

```python
# 5. EKSTRAKSI HASIL CENTROID (Pusat Cluster)
centroids_original_scale = scaler.inverse_transform(kmeans.cluster_centers_)
df_centroid = pd.DataFrame(centroids_original_scale, columns=selected_features)
df_centroid.index = ['Cluster 0', 'Cluster 1']

print("=== DISTRIBUSI ANGGOTA CLUSTER ===")
print(df['Cluster'].value_counts())

print("\n=== PUSAT CLUSTER (NILAI RATA-RATA KARAKTERISTIK) ===")
print(df_centroid)
```

```python
# 6. MELIHAT RELASI CLUSTER DENGAN DIAGNOSE ASLI
cross_tab = pd.crosstab(df['Cluster'], df['Diagnose'])
print("\n=== TABEL SILANG CLUSTER VS DIAGNOSE ===")
print(cross_tab)
```
